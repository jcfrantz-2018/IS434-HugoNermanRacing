```python
# Retrieve metadata from posts

def parse_instafiles(path):

    os.chdir(path)
    
    dataframe = pd.DataFrame(columns=[])
        
    glob('*UTC.json')
    
    for file in glob('*UTC.json'):
        with open(file, 'r') as filecontent:
            filename = filecontent.name[:-5]

            try:
                metadata = orjson.loads(filecontent.read())
            
            except IOError as e:
                print("I/O Error. Couldn't load file. Trying the next one...")
                continue
            else:
                pass
            username = metadata['node']['owner']['username']
            image = metadata['node']['display_url']
            time = datetime.fromtimestamp(int(metadata['node']['taken_at_timestamp']))
            likes = int(metadata['node']['edge_media_preview_like']['count'])   
            
            
            try:
                text = metadata['node']['edge_media_to_caption']['edges'][0]['node']['text']
            except:
                text = ""
            try:
                post_id = metadata['node']['id']
            except:
                post_id = ""
            minedata = {'filename': filename, "username": username, 'time': time, 'text': text,
                    'likes': likes, 'post_id' : post_id, 'image_url':image}
            dataframe = dataframe.append(minedata, ignore_index=True)
            del metadata
            filecontent.close()
    return dataframe
```

```python
os.chdir(os.path.normpath(os.getcwd() + os.sep + os.pardir))
df_rekognition = pd.read_csv('output.csv')
df_rekognition
```

```python
new_filenames = df_rekognition['FileName']
# print(new_filenames)
```

```python
new_list = []
for file in new_filenames:
    filenames = file.split('.', 1)
    appended_filename = filenames[0]
    new_list.append(appended_filename)
    
new_list2 = []
for ch in new_list:
    if ch[-1] != 'C':
        new_str = ch[0:-2]
        new_list2.append(new_str)
    else:
        new_list2.append(ch)
        
# print(new_list2)
```

```python
rekog_words_temp = list(df_rekognition['Rekognition_words'])
rekog_words = []

for words in rekog_words_temp:
    formatted_words = ast.literal_eval(words)
    rekog_words.append(formatted_words)
```

```python

new_df = pd.DataFrame(list(zip(new_list2, rekog_words)), columns=['FileName','Rekognition_words'])
new_df
```

```python
aggregation_functions = {'FileName': 'first', 'Rekognition_words': 'sum'}
new_df2 = new_df.groupby(new_df['FileName']).aggregate(aggregation_functions)
new_df2
```

```python
rekog_list = list(new_df2['Rekognition_words'])
filename_list = list(new_df2['FileName'])

new_rekog = []

for indiv_list in rekog_list:
    myset = set(indiv_list)
    new_rekog.append(list(myset))
```

```python
new_df_rekognition = pd.DataFrame(list(zip(new_list2, instagram_likes, rekog_words)), columns=['FileName','likes','Rekognition_words'])

new_df_rekognition
```

```python
likes = list(new_df_rekognition['likes'])
rekognition_words = list(new_df_rekognition['Rekognition_words'])

word_score_dict = {}

for i in range(len(likes)):
    like_count = likes[i]
    word_list = rekognition_words[i]
    for word in word_list:
        if word in word_score_dict:
            temp_like_count = word_score_dict[word]
            new_like_count = like_count + temp_like_count
            word_score_dict[word] = new_like_count
        else:
            word_score_dict[word] = like_count
            
print(word_score_dict)
```

```python
word_score_df = pd.DataFrame.from_dict(word_score_dict, orient='index', columns=['score'])
sorted_word_score_df = word_score_df.sort_values(by=['score'], ascending=False)

# Top 10 labels
sorted_word_score_df.head(10) 
```
