```python
import json
import os
import pandas as pd
#Firstcolumn -file name , second column, list of the names

files =[]
nameList=[]
df_col2=[]

for filename in os.listdir("C:\\Users\\doyeo\\Desktop\\Hugo\\Hugo Rekognition JSON"):
    if("json" in filename):
        files.append(filename)

        
for file in files:
    f=open(file)
    data= json.load(f)
    for i in range(len(data)):
        nameList.append((data[i]["Name"]))
    
    df_col2.append(nameList)
    nameList=[]
    

df =pd.DataFrame()
df['FileName'] =files
df['Rekognition_words']=df_col2


df


```
